```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=np.array([1,2,3,4,5])
Y=X**2
# Simple Data
plt.plot(X,Y) # simple plot
```

```python
# An object-oriented way of plotting
# Create Data
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
```

```python
fig = plt.figure() # Creates the Empty figure / Canvas
ax = fig.add_axes([0,0,1,1]) # adds the axis and its values
ax.plot(x,y,color="r",label='line1' ) # line Plotted, Line Color Provided, Label for Line is also given
ax.plot(x,z,color="g",label='line2') # Second line plotted on same axis
ax.set_title('Title of the Plot',fontsize=15)    # set title for plot
ax.set_xlabel('Label for X axis',fontsize=15)    # set X label
ax.set_ylabel('Label for Y axis',fontsize=15)    # set Y label
ax.legend() # adds the legend - the color code map for the lines 
ax2 = fig.add_axes([0.10,0.42,0.45,0.45])
```

```python
# This is code for multiple figures
# Create Data
x = np.arange(1,16)   
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
fig = plt.figure(figsize=(8,5)) # empty canvas/figure created
#setting axes as [left, bottom, width, height]
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.10,0.42,0.45,0.45])
#plotting the lines
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax1.legend()    #adding legend
ax2.legend()    #adding legend to plot 2
```

```python
# Creating data
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
fig, (ax1, ax2) = plt.subplots(1, 2) # configuring figure to take two axes - two plots  
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
fig.tight_layout() # to avoid text from y axis of plot 2 conflicting with plot 1
```
