```python
rainfall = pd.read_csv('PythonForDataScienceCodeData/rainfall.csv')
```

```python
rainfall.dropna(inplace = True)
```

```python
rainfall.head()
```

```python
rainfall.info
```

```python
max_district_annual_rainfall = rainfall.groupby('DISTRICT').agg({'ANNUAL': 'max'}).sort_values('ANNUAL', ascending=False)
max_district_annual_rainfall.iloc[0] # returns max annual and which district
```

```python
states_annual = rainfall.groupby('STATE_UT_NAME').agg({'ANNUAL' : 'max'}).sort_values('ANNUAL', ascending=False)
states_annual.head()
```

```python
rainfall.info()
```

```python
rainfall.drop(columns = ['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'], inplace=True)
rainfall.info()
```

```python
rainfall['DISTRICT']
```

```python
pivot_state_mean_table = pd.pivot_table(rainfall.drop('DISTRICT', axis='columns'), index='STATE_UT_NAME', aggfunc='mean')
pivot_state_mean_table
```

```python
rainfall.groupby(['STATE_UT_NAME']).count()[['DISTRICT']]
```

```python
may_max = rainfall.groupby(['STATE_UT_NAME','DISTRICT']).agg({'MAY':'max'}).sort_values('MAY', ascending=False)
may_max.iloc[0]
```
