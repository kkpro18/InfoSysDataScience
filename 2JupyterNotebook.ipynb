```python
#Fuel Efficient
df.loc[(df['mpg'] > 29) & (df['horsepower'] < 93.5) & (df['weight'] < 2500)]
```

```python
# Muscle Cars
df.loc[(df['displacement'] > 262) & (df['horsepower'] > 126) & (df['weight'] > 2799) & (df['weight'] < 3601)]
```

```python
# SUV
df.loc[(df['horsepower'] > 140) & (df['weight'] > 4500)]
```

```python
# Racecar
df.loc[(df['weight'] < 2223) & (df['acceleration'] > 17)]
```

```python
# to sort vehicles by a category
df.sort_values(by = 'cylinders')
```

```python
# low acceleration and high horsepower
df.sort_values(['acceleration', 'horsepower'], ascending = (1,0))
# this sets two values to sort by, default is low - high
# ascending bool is overridden by 1,0 to suggest only acceleration low to high but horsepower not ascending
```

```python
# functions can be applied over specific axis
# to find min and max of the numerical columns e.g cylinder
list1 = [col for col in df.columns if df[col].dtype in ['float', 'int64']]
# the above filters just int/float columns
df[list1].agg(['min', 'max'])
# this grabs the min and max of it.
```

```python
# use groupby to grab a specific subset of a column
df.groupby(['model_year']).count()[['name']]
# groups by model year, and counts it and stores in series name
```

```python
grouped_all = df.groupby(['cylinders','model_year']).agg({'horsepower':['mean','min','max']})
# grabs cylinder, model year and then the min,max,mean for that spec
```

```python
grouped_all.columns = ['hp_mean','hp_min','hp_max']
# better oindg
```

```python
grouped_all = grouped_all.reset_index()
# makes beter indexing
```

```python
grouped_all.head()
# prints 5
```

```python
df.groupby(['model_year']).agg({'acceleration':'mean'}).sort_values('acceleration', ascending = False)[['acceleration']]
# groups by model year and its acceleration in mean, where acceleration is then sorted in descending order
```

```python
# frequency distribution of cylinders for each model year
pd.crosstab(df['model_year'],df['cylinders'])
```

```python
# a pivot table can be used to summarise data stored in a table
pivot1= pd.pivot_table(df[list1], index = 'model_year', aggfunc = np.mean)
pivot1
# this creates a pivot table for means of all numerical data for each model year
```
