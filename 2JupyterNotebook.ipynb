```python
from matplotlib import cm
#creating an empty canvas/figure
fig = plt.figure(figsize=[12,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Scatter plot of Horsepower and Mileage based on Origin",fontsize=16)
ax.set_ylabel('Mileage per gallon',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
# now we can plot based on origins as well to get a more descriptive analysis
origin_list = df["origin"].unique()
colors = cm.jet(np.linspace(0, 1, len(origin_list))) 
for origin, color in zip(origin_list, colors):
    x=df[df["origin"] == origin]["horsepower"]
    y=df[df["origin"] == origin]["mpg"]
    plt.scatter(x,y,color = color, label = origin)
fig.legend()
    
    
```

The highest Horsepower occurs for USA origin vehicles, however it has a low mileage.
The highest MPG is Japan origin vehicles.

To know numbers of cars released in each year, a bar chart may be suitable.
Bar Chart can be created using ax.scatter(x, height, width, bottom, align)

It can be seen that the most cars were produced in the year of 1973 with value 39.

The least number of cars were  produced in 1974 with value 26.

Histogram stores a frequency distribution of a given data.


```python
# to find the distribution of horsepower data
fig = plt.figure(figsize=[8,6])
ax = fig.add_axes([0,0,1,1])
ax.hist(df['horsepower'], bins=20)
ax.set_title("Distribution of Horsepower")
ax.set_ylabel("Frequency")
ax.set_xlabel("Horsepower")
fig.show()
```

Pie Chart can be created using
ax.pie(x, labels)
x = wedge sizes, a 1-d array
labels = list of labels for wedges

```python
# do a pie chart for distribution based on origin of vehicles
pie_df = pd.DataFrame()    #creating a sub dataframe to plot the pie chart
pie_df['Count'] = df['origin'].value_counts()    #getting the count of ‘origin’ and assigning to the df
pie_df = pie_df.reset_index()    #re-arranging the index
pie_df.rename(columns={'origin':'Country'},inplace=True) #re-naming the col
pie_df
```

```python
fig = plt.figure(figsize=[8,6])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
explode = [0.05, 0, 0] # elevated pies
colors = ['red', 'pink', 'blue' ]
ax.set_title('Origin of the Cars')
#plotting pie chart
# ax.pie(pie_df['Count'], labels=pie_df['Country'])
ax.pie(pie_df['Count'], labels=pie_df['Country'],
        autopct='%0.1f%%',explode = explode, 
        shadow = True, startangle = 250,
        colors = colors)
ax.legend()
fig.show()
```

Most vehicles originate from USA, where the minority comes from europe.

To view historic variations and trends, a line chart may be used.


```python
# Generate a report comparing, mileage, horsepower and weight of cars in different years
line_df = df[list1].groupby(by='model_year')
line_df = line_df.mean()
line_df
```

```python
fig = plt.figure(figsize=[8,4])
ax = fig.add_axes([0,0,1,1])
ax.plot(line_df['mpg']/10, label="MPG (1/10)")
ax.plot(line_df['weight']/1000, label="Weight (1/1000)")
ax.plot(line_df['horsepower']/100, label="Horsepower (1/100)")

ax.set_title("Comparison between Mileage, Horsepower and Weight")
ax.set_xlabel("Model Year")
ax.set_ylabel("Value")
ax.legend()
fig.show()
```

```python
fig = plt.figure(figsize=[8,4])
ax = fig.add_axes([0,0,1,1])
# line styles can be applied to make the lines different e.g -,--,:,-.
ax.plot(line_df['mpg']/10, label="MPG (1/10)", linestyle='-.')
ax.plot(line_df['weight']/1000, label="Weight (1/1000)", linestyle='--')
ax.plot(line_df['horsepower']/100, label="Horsepower (1/100)", linestyle=':')

ax.set_title("Comparison between Mileage, Horsepower and Weight")
ax.set_xlabel("Model Year")
ax.set_ylabel("Value")
ax.legend()
fig.show()

# texts and annotation can be added
plt.text(75.5, 1.14, 'Horsepower',
             va = 'center', #va -> vertical alignment
             rotation = 4, #angle
             bbox = dict(boxstyle = 'round', facecolor = 'wheat', alpha = 0.3)) #background style
plt.text(75.5, 2.25, 'Mileage',va = 'center',rotation = 17,bbox = dict(boxstyle = 'round',facecolor = 'wheat',alpha = 0.3))
plt.text(76, 3.15, 'Weight',va = 'center',rotation = -13,bbox = dict(boxstyle = 'round',facecolor = 'wheat',alpha = 0.3))
# the three plt.text above shows the label above the centre of the line
plt.annotate('Highest Mileage', xy=(80.1, 3.39), xytext=(80.5, 3.38), arrowprops=dict(facecolor='wheat', shrink=.01))
plt.annotate('Lowest Mileage', xy=(73.05, 1.70), xytext=(73.5, 1.69), arrowprops=dict(facecolor='wheat', shrink=.01))
# the above shows an arrow pointing to the highest and lowest point of the mileage.
```

Same plot but different line style and labels for each line + a arrow at lowest and highest point

```python
fig = plt.figure(figsize=[8,4])
ax = fig.add_axes([0,0,1,1])
# line styles can be applied to make the lines different e.g -,--,:,-.
ax.plot(line_df['mpg']/10, label="MPG (1/10)", linestyle='-.', marker='s')
ax.plot(line_df['weight']/1000, label="Weight (1/1000)", linestyle='--', marker='o')
ax.plot(line_df['horsepower']/100, label="Horsepower (1/100)", linestyle=':', marker='+')

ax.set_title("Comparison between Mileage, Horsepower and Weight")
ax.set_xlabel("Model Year")
ax.set_ylabel("Value")
ax.legend()
fig.show()
```

Same plot as the last two plots but this one has marker symbols at each point to make the appearance more evident.
The highest MPG was recorded in 1980, and lowest in 1973.
The heaviest vehicles were produced in 1970, where it gets lighter over the years where the lowest is at 1982.
The highest horsepower vehicles were produced in 1970, whereas over the years it has dropped.
