```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x) * np.exp(-x)
fig = plt.figure(figsize=[6,6])
ax = fig.add_axes([0,0,1,1])
ax.set_title("Main Heading", fontsize=16)
fig.suptitle("Figure Title", fontsize=12)
ax.plot(x,y,color = 'black'
        , marker="o", markerfacecolor='cyan')
ax.set_xlabel("Abscissa", color="g")
ax.set_ylabel("Ordinate", color="g")
plt.annotate('2nd crest', xy=(2,0.165), xytext=(2.35,0.25), arrowprops=dict(facecolor='#196F3D', shrink=.01))
#1,0.39, 0=60°
plt.annotate('θ = 60°',xy=(1,0.39))
```
